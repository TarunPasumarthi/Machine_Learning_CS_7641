### 3.2 Poisson distribution [5pts]

The Poisson distribution is defined as 

$$P(X=k)=\frac{\lambda^k e^{-\lambda}}{k!} (k=0,1,2,...).$$

What is the maximum likelihood estimator of $\lambda$?(Suppose we have $n$ i.i.d data points $x_1,\dots,x_n$).

* <span style="color:red">
        $L(\lambda)= log \prod_{i = 1}^{n} \frac{\lambda^k e^{-\lambda}}{k!}\\
        =\sum_{i = 1}^{n} log(\frac{\lambda^k e^{-\lambda}}{k!})\\
        =\sum_{i = 1}^{n} log(e^{-\lambda})+log(\lambda^k)-log(k!)\\
        =\sum_{i = 1}^{n} -\lambda + k+log(\lambda) -log(k!)\\
        \frac{\partial log(L(\theta))}{\partial \theta}=-n + \frac{1}{\lambda}\sum_{i = 1}^{n} k\\
        -n + \frac{1}{\lambda}\sum_{i = 1}^{n} k=0\\
        \lambda=\frac{1}{n}\sum_{i = 1}^{n} k
        $
    </span>
